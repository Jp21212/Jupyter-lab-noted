```python
CSS (Cascading style sheets) is a language used to describe the presentation of a web page written in HTML. It
controls the layout, color, fonts, and overall visual appearence of web pages.

CSS syntax: A CSS rule consist of a selector and a declration block
Selector {
    property: value;
}

1. Selector: specifies the HTML elements you want to style
2. Property: The style attribute you want to change
3. Value: The value of the property

Ex:
P {
color: blue
    font-size: 14px
}
This styles ass <p> elements to have blue text and a font-size of 14 pixels

2. Adding CSS to HTML using external CSS

External CSS
Uses an external .css file linked with the <link> tag.
<head>
<link rel="stylesheet" href="style.css">
</head>

3. CSS. Selectors
Element selector
Selects all elments of a given type

h1 {
color:red;
}

Class Selector
Selects elements wwith specific clas attribute. Uses a period . followed by the class name

.intro {
font-size: 16px
}

A class selector must be paired with creating the class in HTML file

<p> class="intro"> This is an introductory paragraph. <p>

Id Selector
Selects an element with a specifc class attribute. Use a period . followed by the class name. 

#main {
width: 960px;
}

An Id selectorr must be paired with creating the Id in an HTML fule. An ID is used to select a very specific HTML
element. We can use them with classes to further differentiate among elements

<div id="main">This is the main content area </div>

Universal Selector

Selects all elements
* {
margin: 0;
padding: 0;
}

Grouping selectors
Applies the same sytles to multipale selectors

h1, h2, h3 {
color: navy;
}

4. Common CSS properties

Text properties
Color: Sets the color of the text

p{
color: #333333
}

font-size: Sets the size of the font

p{
font-size: 14 px
}

font-family: 
specifies the font of the text

p {
font-family: Arial, sans-serif;
}

font-wieght: Sets the weight (boldness) of the font

p {
font-wieght: bold;
}

text-align: Aligns the text horizontally

p{
text-align: center
}

Background properties
Background-color: Sets the background color of an element

body {
background-color: f0f0f0;
}

Background-image: Sets a background image fo an element

body{
background-image: url('background.jpg');
}

Box Model properties
The CSS box model consist of margins, boarders, padding, and the content area.

width and hieght: Sets the size of the content area.

div {
width: 200px;
height: 100px;
}

padding: sets the space between the content and the boarder

div{
padding: 10px;
}

boarder: sets the boarder around an element

div{
border: 1px solid #000;
}

Margin: Sets the space outside the boarder

div{
margin: 20px;
}

Display and Visability
display: Specifies the display behavior of an element

.inline {
display: inline;
}

.block{
display: block;
}

.none {
display: none; /* Hides the element */
}

visability: controls the visability of an element 

.invisable { 
visability : hidden; /* Element takes up space but is not visible 
*/
}

5. CSS Flexbox

Flexbox is a layout mode in CSS designed to provide a more efficent way to lay out, align, and distrubute space
among items in a container

Setting up a flex container
. container {
display: flex;
}

Flex direction
Specifies the direction of the flex items . 

.container {
flex-direction: row; /* row, row-reverse, column, column-reverse */
}

Justify content
align items along the main axis.

.container {
    Justify-content: center; /* flex-start, flex-end, center, space-between, space-around */
}

Align items 
Align items along the cross axis
.container {
    align-items: center; /*stretch, flex-start, flex-end, center, baseeline */
}

Flex Wrap
Allows flex items to wrap onto multiple lines. 

.container {
    flex-wrap: wrap; /* nowrap, wrap-reverse */ 
}

Flex Items properties

flex-grow: Specifies how much a flex item will grow relative to the rest

.item {
    flex-grow: 1;
}

flex-shrink: Specifies how much a flex item will shrink relative to the rest.

.item {
    flex-shrink: 1;
}

Example FLEXBOX layout
HTML:

<div class="container">
<div class="item">Item 1</div>
<div class="item">Item 2</div>
<div class="item">Item 3</div>
</div>

CSS

.container { 
display: flex;
flex-direction: row;
justify-content: space-between;
align-items: center;
}

.item {
    background-color: lightblue;
    padding: 20px;
    margin: 10px;
    flex: 1;
}

This creates a responsive layout where items are evenly spaced and aligned. 
```
